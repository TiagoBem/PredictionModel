```python
#########################
#  Real PreProcessing  ##
#########################

import glob
import pandas as pd
import numpy as np
from datetime import datetime

def getWeekday(timestamp):
  return timestamp.dayofweek

def getMonth(timestamp):
  return timestamp.month

def getHours(timestamp):
  return timestamp.hour

def parseToDate(date):
  return datetime.strptime(date, '%Y.%m.%d_%H.%M.%S')

def parseToSeconds(date):
  return (date - datetime(1970,1,1)).total_seconds()

dataset = pd.DataFrame([])
totalrows = 0
all_filenames = [i for i in glob.glob('datafiles/*.{}'.format('csv'))]
for f in all_filenames:
  dataframe = pd.read_csv(f)
  totalrows += len(dataframe.index)
  dataframe.pop('Longitude')
  dataframe.pop('Latitude')
  dataframe.pop('Operatorname')
  dataframe.pop('CellID')
  dataframe.pop('NetworkMode')
  dataframe.pop('SNR')
  dataframe.pop('CQI')
  dataframe.pop('RSSI')
  dataframe.pop('State')
  dataframe.pop('NRxRSRP')
  dataframe.pop('NRxRSRQ')
  dataframe.pop('ServingCell_Lon')
  dataframe.pop('ServingCell_Lat')
  dataframe.pop('ServingCell_Distance')

  dataframe['Timestamp'] = dataframe['Timestamp'].apply(parseToDate)

  dataframe['Weekday'] = dataframe['Timestamp'].apply(lambda row: getWeekday(row))
  dataframe['Month'] = dataframe['Timestamp'].apply(lambda row: getMonth(row))
  dataframe['Hours'] = dataframe['Timestamp'].apply(lambda row: getHours(row))
  dataframe['TimeDiff'] = dataframe['Timestamp'].apply(lambda row: parseToSeconds(row)) - dataframe['Timestamp'].shift().apply(lambda row: parseToSeconds(row))

  column = dataframe.pop('Month')
  column2 = dataframe.pop('Weekday')
  column3 = dataframe.pop('Hours')
  dataframe.insert(1,'Month', column)
  dataframe.insert(1,'Weekday', column2)
  dataframe.insert(1,'Hours', column3)

  dataframe.pop('Timestamp')

  dataframe['Speed_N-1'] = dataframe['Speed'].shift()
  dataframe['RSRP_N-1'] = dataframe['RSRP'].shift()
  dataframe['RSRQ_N-1'] = dataframe['RSRQ'].shift()
  dataframe['RSRP_N-1'] = dataframe['RSRP'].shift()
  dataframe['DL_bitrate_N-1'] = dataframe['DL_bitrate'].shift()
  dataframe['UL_bitrate_N-1'] = dataframe['UL_bitrate'].shift()

  dataframe['Speed_N-2'] = dataframe['Speed'].shift().shift()
  dataframe['RSRP_N-2'] = dataframe['RSRP'].shift().shift()
  dataframe['RSRQ_N-2'] = dataframe['RSRQ'].shift().shift()
  dataframe['RSRP_N-2'] = dataframe['RSRP'].shift().shift()
  dataframe['DL_bitrate_N-2'] = dataframe['DL_bitrate'].shift().shift()
  dataframe['UL_bitrate_N-2'] = dataframe['UL_bitrate'].shift().shift()

  dataframe['Speed_N-3'] = dataframe['Speed'].shift().shift().shift()
  dataframe['RSRP_N-3'] = dataframe['RSRP'].shift().shift().shift()
  dataframe['RSRQ_N-3'] = dataframe['RSRQ'].shift().shift().shift()
  dataframe['RSRP_N-3'] = dataframe['RSRP'].shift().shift().shift()
  dataframe['DL_bitrate_N-3'] = dataframe['DL_bitrate'].shift().shift().shift()
  dataframe['UL_bitrate_N-3'] = dataframe['UL_bitrate'].shift().shift().shift()

  dataframe['Speed_N-4'] = dataframe['Speed'].shift().shift().shift().shift()
  dataframe['RSRP_N-4'] = dataframe['RSRP'].shift().shift().shift().shift()
  dataframe['RSRQ_N-4'] = dataframe['RSRQ'].shift().shift().shift().shift()
  dataframe['RSRP_N-4'] = dataframe['RSRP'].shift().shift().shift().shift()
  dataframe['DL_bitrate_N-4'] = dataframe['DL_bitrate'].shift().shift().shift().shift()
  dataframe['UL_bitrate_N-4'] = dataframe['UL_bitrate'].shift().shift().shift().shift()

  dataframe = dataframe.iloc[4:]

  filter = (dataframe['TimeDiff'] >= 0)
  dataframe = dataframe.loc[filter]  

  dataset = pd.concat([dataset, dataframe])

dataset = dataset.replace('-',np.NaN)
dataset = dataset.dropna()
dataset.to_csv('NewDataset.csv')
print(dataset)
print(totalrows)
```
